### kept confirmed case data in a csv file and transferred death data to database, now want to read all in one dataframe and start to analysis that

```python
# read death data from database (from a view in database)
# i created death_all_groupbycountry view in database . the view includes data group by country .
death_all_groupbycountry = pd.read_sql_query('select * from death_all_groupbycountry', con=engine)
death_all_groupbycountry.head(100)
```

```python
#merge confirmed case and death data in one dataframe
df = pd.merge(confirmed_all_groupbycountry,death_all_groupbycountry,on='Country/Region')
df.head(100)
```

```python
df = df.reindex(sorted(df.columns), axis=1)
df.head()
```

```python
cols = list(df.columns.values)
cols
```

### the most number of death in last day of data

```python
lastday_df = df[['Country/Region', '4/22/2020_x','4/22/2020_y']]
lastday_df = lastday_df.sort_values(by='4/22/2020_y',ascending=False)
lastday_df.head(100)
```

### the most number of death in last day of data 

```python
most_death_country_df= lastday_df.head(20)
title = 'number of death on 04/22/2020 '
ax = most_death_country_df.plot(kind='bar',x='Country/Region',y=['4/22/2020_x','4/22/2020_y'],title=title)
ax.set_xlabel("country")
ax.set_ylabel("number of case and death")
ax.legend(["number of case", "number of death"]);
```
